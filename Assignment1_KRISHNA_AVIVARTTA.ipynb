```python
gdp_per_capita = df2['GDP'].values / df2['Population'].values
print(gdp_per_capita)
```

Using SkLearn

```python
from sklearn.linear_model import LinearRegression
x_1d = gdp_per_capita
X = gdp_per_capita.reshape(-1, 1)
Y = life_satisfaction

model = LinearRegression().fit(X,Y)
print(f"intercept(m): {model.intercept_}") 
print(f"slope(c): {model.coef_}")
```

```python
regressor = LinearRegression()
regressor.fit(X, Y)

y_pred = regressor.predict(X)

sns.scatterplot(x= x_1d, y= Y, color = 'blue')
sns.scatterplot(x= x_1d, y=y_pred, color = 'red')
```
