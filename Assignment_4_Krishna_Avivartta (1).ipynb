```python
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd

# machine learning models
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

# machine learning metrices
from sklearn.metrics import accuracy_score, classification_report, plot_confusion_matrix
```

```python
from sklearn.datasets import load_iris
df = load_iris()
features = df['data']
feature_names = df['feature_names']
target = df['target']
```

```python
# Two large arrays are defined, features and target.
# The raw data about the irises is in features; features is a 2D array.

features.shape
```

```python
# segregating features and target class
X = df.data 
y = df.target
```

```python
# Creating Train and Test datasets
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 50, test_size = 0.25)

# for the ease of producing the classification reports
def report(model):
    preds = model.predict(X_test)
    print(classification_report(preds,y_test))
    plot_confusion_matrix(model,X_test,y_test)
```

K- Nearest Neighbors

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)

report(knn)
```

Decision Trees

```python
#model fitting and classification
dtc_model = DecisionTreeClassifier()
dtc_model.fit(X_train,y_train)
y_pred = dtc_model.predict(X_test)

import matplotlib.pyplot as plt
from sklearn import tree
# plotting
fig = plt.figure(figsize=(25,20))
plot = tree.plot_tree(dtc_model, filled=True)

report(dtc_model)
```

Naive Bayes

```python
#model fitting and classification
nb = GaussianNB()
nb.fit(X_train,y_train)
y_pred = nb.predict(X_test)

report(nb)
```

SVM - Support Vector Machine

```python
svm = SVC()
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)

report(svm)
```
